```python
plt.bar(df['time'],df['total_bill'],color="g")
plt.title("BAR GRAPH OF TOTAL BILL VS TIME")
plt.xlabel("TIME")
plt.ylabel("TOTAL BILL")
plt.show()
```

```python
plt.bar(df['day'],df['total_bill'],color="b")
plt.title("BAR GRAPH OF TOTAL BILL VS DAY")
plt.xlabel("DAY")
plt.ylabel("TOTAL BILL")
plt.show()
```

```python
plt.bar(df['day'],df['tip'],color= "purple")
plt.title("BAR GRAPH OF TIP VS DAY")
plt.xlabel("DAY")
plt.ylabel("TIP")
plt.show()
```

```python
sns.boxplot(x="time",y="total_bill",data=df)
plt.show()
```

```python
sns.boxplot(x="time",y="tip",data=df)
plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.hist(df['total_bill'],color = 'g',rwidth =0.95)

plt.title('HISTOGRAM OF TOTAL BILL',size = 15,color ='black')
plt.xlabel('TOTAL BILL')
plt.ylabel('COUNT')
plt.show()
```

```python
plt.figure(figsize=(2,8))
plt.hist(df['smoker'],color = 'yellow',rwidth =0.98)
plt.title('HISTOGRAM OF SMOKERS',size = 15,color ='BLACK')
plt.xlabel('SIZE')
plt.ylabel('COUNT')
plt.show()
```
